```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
conn = sqlite3.connect('airbnb.db')
```

```python
data_path = '/content/AB_NYC_2019.csv'
df = pd.read_csv(data_path)
```

```python
df.to_sql('listings', conn, if_exists='replace', index=False)
```

```python
cursor = conn.cursor()

# Execute the SQL query
cursor.execute("SELECT * FROM listings LIMIT 10;")

# Fetch the results
results = cursor.fetchall()

# Display the first 10 rows
for row in results:
    print(row)
```

```python
cursor.execute("""
    SELECT room_type, COUNT(*) AS num_listings
    FROM listings
    GROUP BY room_type;
""")
results = cursor.fetchall()
```

```python
for row in results:
    print(row)
```

```python
cursor.execute("""
    SELECT AVG(number_of_reviews) AS avg_reviews
    FROM listings;
""")

# Completed
result = cursor.fetchone()
# Print the average number of reviews per listing
print("Average number of reviews per listing:", result[0])
```

```python
cursor.execute("""
    SELECT id
    FROM listings
    WHERE number_of_reviews = (
        SELECT MIN(number_of_reviews)
        FROM listings
    )
    LIMIT 15;
""")

# Completed
results = cursor.fetchall()
```

```python
for row in results:
    print(row)
```

```python
cursor.execute("""
    SELECT neighbourhood, price
    FROM listings
    WHERE neighbourhood = 'Chelsea'
      AND price <= 200;
""")
results = cursor.fetchall()

# Display results
for row in results:
    print(row)
```
